```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import sys
import math
```

# Aufgabe 1a

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)


F1 = np.sqrt(x)
F2 = (np.log(x))/x+1


ax.plot(x, F1, color="blue", label="$y_1(x)$")
ax.plot(x, F2, color="green", label="$y_1(x)$")

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.legend()

plt.show()

```

# Aufgabe 1b

```python
fig, ax = plt.subplots(1,2, figsize=(12,3))
x = np.linspace(0, 10, 100)


F1 = np.exp(x)

ax[0].set_title('Linear')
ax[1].set_title('SemiLog')

ax[0].plot(x,F1,color = "blue", label = "$y_1(x)$")
ax[1].semilogy(x, F1, color = "green", label="$y_1(x)$")



for axis in ax:
    axis.set_xlabel('X-Axis')
    axis.set_ylabel('Y-Axis')
    axis.legend()

plt.show()

```

# Aufgabe 1c

```python
fig, ax = plt.subplots(1,3, figsize=(12,3))

x = np.linspace(0, 10, 100)
n = np.linspace(-100, 100, 10)


F1 = x**(1/2)   # np.sqrt(x)


ax[0].plot(x,F1,color = "blue", label = "$y_1(x)$")
ax[1].loglog(x,F1,color = "green", label = "$y_1(x)$")




for number in n: 
    F2 = x**number
    ax[2].loglog(x,F2,color = "pink", label = "$y_1(x)$")
   

    ax[0].set_title('Linear')
    ax[1].set_title('LogLog')
    ax[2].set_title('LogLog(f(x) = x**n)')


for axis in ax:
    axis.set_xlabel('X-Axis')
    axis.set_ylabel('Y-Axis')
    


plt.show()

```

```python
fig, ax = plt.subplots(1,2, figsize=(12,3))

x = np.linspace(0, 10, 10)
y = np.linspace(0, 10, 10)
t = np.linspace(-3,3,10)
a = 0.1
e = 5


b = a*(np.sqrt((e**2) - 1))

F1 = x + a*(np.cosh(t))
F2 = b*(np.sinh(t))


ax[0].plot(x,F1,color = "blue", label = "$y_1(x)$")
ax[1].plot(x,F2,color = "green", label = "$y_1(x)$")



for axis in ax:
    axis.set_xlabel('X-Axis')
    axis.set_ylabel('Y-Axis')
    


plt.show()

```
